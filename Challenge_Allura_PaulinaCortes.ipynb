### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 6. Gráficos (ANEXOS)



```python
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

ingresos = [
    ingreso_tienda1,
    ingreso_tienda2,
    ingreso_tienda3,
    ingreso_tienda4
]

valoraciones = [
    valoracion_t1,
    valoracion_t2,
    valoracion_t3,
    valoracion_t4
]

envios = [
    envio_promedio_t1,
    envio_promedio_t2,
    envio_promedio_t3,
    envio_promedio_t4
]

colores_pastel = ['#CDB4DB', '#FFC8DD', '#BDE0FE', '#A2D2FF']
```

```python
def grafico_ingresos():
    plt.bar(tiendas, ingresos, color=colores_pastel)
    plt.title('Ingreso total por tienda')
    plt.xlabel('Tienda')
    plt.ylabel('Ingreso total')
    plt.show()

def grafico_proporcion():
    plt.pie(
        ingresos,
        labels=tiendas,
        autopct='%1.1f%%',
        startangle=90,
        colors=colores_pastel
    )
    plt.title('Proporción del ingreso total por tienda')
    plt.axis('equal')
    plt.show()

def grafico_calificacion():
    plt.plot(tiendas, valoraciones, marker='o', color='#9B5DE5')
    plt.title('Calificación promedio por tienda')
    plt.xlabel('Tienda')
    plt.ylabel('Calificación promedio')
    plt.show()

def grafico_envio():
    plt.plot(tiendas, envios, marker='o', color='#9B5DE5')
    plt.title('Costo de envío promedio por tienda')
    plt.xlabel('Tienda')
    plt.ylabel('Costo de envío promedio')
    plt.show()

def grafico_relacion():
    sizes = [i / 5_000_000 for i in ingresos]
    plt.scatter(valoraciones, envios, s=sizes, c=colores_pastel)
    for i, tienda in enumerate(tiendas):
        plt.text(valoraciones[i], envios[i], tienda)
    plt.title('Relación entre valoración, costo de envío e ingresos')
    plt.xlabel('Calificación promedio')
    plt.ylabel('Costo de envío promedio')
    plt.show()

grafico_ingresos()
grafico_proporcion()
grafico_calificacion()
grafico_envio()
grafico_relacion()
```

## ¡Extra! Análisis del desempeño geográfico

En este análisis se utilizaron las coordenadas de latitud y longitud para visualizar la distribución geográfica de las ventas de las cuatro tiendas. El objetivo fue identificar posibles zonas con mayor concentración de ventas y analizar si la ubicación influye en el desempeño de cada tienda.

El gráfico muestra que las ventas se concentran en regiones similares para todas las tiendas, sin que alguna destaque claramente por dominar una zona específica. Las áreas con mayor densidad de puntos representan regiones con mayor actividad de compra.

Aunque existen ventas de mayor valor en algunas zonas, estas diferencias no explican por sí solas el desempeño general de las tiendas. En conjunto, el análisis sugiere que la ubicación geográfica no es un factor determinante y que las diferencias entre tiendas están más relacionadas con factores económicos y comerciales.



```python
import pandas as pd
import matplotlib.pyplot as plt

df_geo = pd.concat([
    tienda.assign(Tienda='Tienda 1'),
    tienda2.assign(Tienda='Tienda 2'),
    tienda3.assign(Tienda='Tienda 3'),
    tienda4.assign(Tienda='Tienda 4')
])
```

```python
plt.figure(figsize=(8,6))

for nombre in df_geo['Tienda'].unique():
    df_temp = df_geo[df_geo['Tienda'] == nombre]
    plt.scatter(
        df_temp['lon'],
        df_temp['lat'],
        s=df_temp['Precio'] / 5000,
        alpha=0.3,
        label=nombre
    )

plt.title('Distribución geográfica de ventas por tienda')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.legend()
plt.show()
```
